```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap

### demand projection assumptions
def demand_proj_assumption(file_path, run_name, output_folder, called = False):
    if called == True:
        User_cons = pd.read_csv(file_path + 'User_con_' + run_name + '.csv', sep = ',')
        demand_proj =  User_cons[User_cons['7'].str.contains('rep')]
        demand_proj['group'] = demand_proj['2'].astype(str).str.cat(demand_proj['7'].astype(str), sep='_')
        demand_proj['demand'] = demand_proj.groupby('group')['8'].transform('sum')/(10**3)
        # demand_proj = demand_proj[~demand_proj['group'].duplicated()]

        list_rep = demand_proj['7'].unique()
        for rep in list_rep:
            rep_proj = demand_proj[demand_proj['7'] == rep]
            rep_proj['8'] = rep_proj['8']/(10**3)

            values1 = np.linspace(0, 1, 20)
            values2 = np.linspace(0, 1, 9)
            # Create two colormaps
            cmap1 = plt.cm.tab20
            cmap2 = plt.cm.Set3


            colors1 = cmap1(values1)
            colors2 = cmap2(values2)
            # cmap = plt.cm.get_cmap('tab20', 20)
            # cmap1 = plt.cm.get_cmap('Set3', 9)


            # Combine the colors from both colormaps
            combined_colors = np.vstack((colors1, colors2))

            # Create a new colormap using the combined colors
            combined_cmap = ListedColormap(combined_colors)

            pivot_df = rep_proj.pivot(index='3', columns='4', values='8')
            column_sums = pivot_df.sum(skipna=True)

            # Step 2: Sort the columns based on their sums
            sorted_columns = column_sums.sort_values(ascending=False)

            # Step 3: Reindex the DataFrame with the sorted column order
            sorted_df = pivot_df.reindex(sorted_columns.index, axis=1)
            sorted_df.head()
            # Plot with the discrete color map
            sorted_df.plot(kind='bar', stacked=True, colormap=combined_cmap)
            plt.xlabel('Year')
            plt.xticks(rotation = 0)
            plt.ylabel('$ Billions')
            plt.title(f'Projection {rep}')
            plt.legend(title='', bbox_to_anchor=(0.5, -0.6), loc = 'lower center', ncol=4)

            plt.savefig(output_folder + 'demand_projection_Kinesys.pdf'
                        , format ='pdf'
                        ,  bbox_inches='tight')
        # plt.show()
    else:
        pass
```
